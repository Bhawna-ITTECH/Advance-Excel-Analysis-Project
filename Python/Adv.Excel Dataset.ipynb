```python
# Loan EMI validate karo (given Loan Amount, Rate, Tenure).
def EMI(P, rate, tenure_months):
    r = rate / (12*100)
    n = tenure_months
    emi = P * r * (1+r)**n / ((1+r)**n - 1)
    return emi
# Example:
emi_value = EMI(500000, 12, 60)
emi_value
```
